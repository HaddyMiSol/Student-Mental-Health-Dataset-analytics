```python
student_df['Choose your gender'].unique()
```

```python
#Replace duplicate values 
student_df['What is your course?'].unique()
```

```python
student_df['What is your course?'] = student_df['What is your course?'].replace(['Pendidikan Islam ','Pendidikan islam'],'Pendidikan Islam')
student_df['What is your course?'] = student_df['What is your course?'].replace(['Islamic education'],'Islamic Education')
student_df['What is your course?'] = student_df['What is your course?'].replace(['Diploma Nursing','Nursing '],'Nursing')
student_df['What is your course?'] = student_df['What is your course?'].replace(['koe','Koe'],'KOE')
student_df['What is your course?'] = student_df['What is your course?'].replace(['psychology'],'Psychology')
student_df['What is your course?'] = student_df['What is your course?'].replace(['Laws'],'Law')
student_df['What is your course?'] = student_df['What is your course?'].replace(['engin','Engine'],'Engineering')
student_df['What is your course?'] = student_df['What is your course?'].replace(['Fiqh','Fiqh fatwa '],'Fiqh fatwa')
student_df['What is your course?'] = student_df['What is your course?'].replace(['Benl'],'BENL')
student_df['What is your course?'] = student_df['What is your course?'].replace(['Kirkhs'],'KIRKHS')
```

```python
student_df['What is your course?'].unique()
```

```python
#Replace 'year 1' with '1'
student_df['Your current year of Study'].unique()
```

```python
student_df['Your current year of Study'] = student_df['Your current year of Study'].replace(['year 1'],'Year 1')
student_df['Your current year of Study'] = student_df['Your current year of Study'].replace(['year 2'],'Year 2')
student_df['Your current year of Study'] = student_df['Your current year of Study'].replace(['year 3'],'Year 3')
student_df['Your current year of Study'] = student_df['Your current year of Study'].replace(['year 4'],'Year 4')
```

```python
student_df['Your current year of Study'].unique()
```

```python
#Group into section 
student_df['What is your CGPA?'].unique()
```

```python
student_df['What is your CGPA?'] = student_df['What is your CGPA?'].replace(['3.50 - 4.00 '],'3.50 - 4.00')
```

## Exploratory Analysis and Visualization





### Mental Health Disaggregation

```python
Depr = student_df['Do you have Depression?'].value_counts()
Depr
Anx = student_df['Do you have Anxiety?'].value_counts()
Anx
Panic_A = student_df['Do you have Panic attack?'].value_counts()
Panic_A
Treatment= student_df['Did you seek any specialist for a treatment?'].value_counts()
```

```python
Depr_by_sex = student_df[student_df['Do you have Depression?']=='Yes']['Choose your gender'].value_counts()
Depr_by_sex
Anx_by_sex = student_df[student_df['Do you have Anxiety?']=='Yes']['Choose your gender'].value_counts()
Anx_by_sex
Panic_by_sex = student_df[student_df['Do you have Panic attack?']=='Yes']['Choose your gender'].value_counts()
Panic_by_sex
Treat_by_sex = student_df[student_df['Did you seek any specialist for a treatment?']=='Yes']['Choose your gender'].value_counts()
Treat_by_sex
```

```python
fig, axes = plt.subplots(2,4, figsize=(16, 8))
sns.set_palette(sns.color_palette('Set1'))

axes[0,0].set_title("Do you have Depression?", size=14, x = 0.52, y =0.95)
axes[0,0].pie(Depr, labels = Depr.index, autopct = '%1.1f%%', startangle = 45);

axes[0,1].set_title("Do you have Anxiety?", size=14, x = 0.52, y =0.95)
axes[0,1].pie(Anx, labels = Anx.index, autopct = '%1.1f%%', startangle = 45);

axes[0,2].set_title("Do you have Panic Attack?", size=14, x = 0.52, y =0.95)
axes[0,2].pie(Panic_A, labels = Panic_A.index, autopct = '%1.1f%%', startangle = 45);

axes[0,3].set_title("Did you seek any specialist for a treatment?", size=14, x = 0.52, y =0.95)
axes[0,3].pie(Treatment, labels = Treatment.index, autopct = '%1.1f%%', startangle = 45);

#sns.set_palette(sns.color_palette('rainbow_r'))
axes[1,0].set_title("Depression by gender", size=14, x = 0.52, y =0.95)
axes[1,0].pie(Depr_by_sex, labels = Depr_by_sex.index, autopct = '%1.1f%%', startangle = 45);

axes[1,1].set_title("Anxiety by gender", size=14, x = 0.52, y =0.95)
axes[1,1].pie(Anx_by_sex, labels = Anx_by_sex.index, autopct = '%1.1f%%', startangle = 45);

axes[1,2].set_title("Panic by gender", size=14, x = 0.52, y =0.95)
axes[1,2].pie(Panic_by_sex, labels = Panic_by_sex.index, autopct = '%1.1f%%', startangle = 45);

axes[1,3].set_title("Treatment by gender", size=14, x = 0.52, y =0.95)
axes[1,3].pie(Treat_by_sex, labels = Treat_by_sex.index, autopct = '%1.1f%%', startangle = 45);
```

### Depressed/ Anxiety/ Panic attack by Age group 

```python
#Create a graph that shows Depression,Anxiety, Panic Attack by age group
def get_num_people_by_age_category(df):
    student_df["age_group"] = pd.cut(x=student_df['Age'], bins=[10,15,20,25,30,40,100], labels=["10-15yrs","15-20yrs","20-25yrs","25-30yrs","30-40","Older"])
    return df

# Call function
df = get_num_people_by_age_category(student_df)
#df
```

```python
#Age group of those depressed
Depr_Yes=df[df['Do you have Depression?']=='Yes']
Depr_Age= Depr_Yes[['Do you have Depression?','age_group']]

#Age group of those with anxiety
Anx_Yes=df[df['Do you have Anxiety?']=='Yes']
Anx_Age= Anx_Yes[['Do you have Anxiety?','age_group']]

#Age group of those with panic attack
Panic_Yes=df[df['Do you have Panic attack?']=='Yes']
Panic_Age= Panic_Yes[['Do you have Panic attack?','age_group']]
```

```python
#value count by age group
Depr_Age= Depr_Age['age_group'].value_counts()

Anx_Age= Anx_Age['age_group'].value_counts()

Panic_Age= Panic_Age['age_group'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
plt.title('Age group of depressed students')
plt.xlabel('Age')
plt.ylabel('No Depressed')

sns.barplot(x=Depr_Age.index, y= Depr_Age);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Age group of students with anxiety')
plt.xlabel('Age')
plt.ylabel('Anxiety')

sns.barplot(x=Anx_Age.index, y= Anx_Age);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Age group of students with panic attack')
plt.xlabel('Age')
plt.ylabel('No Panic')

sns.barplot(x=Panic_Age.index, y= Panic_Age);
```

## Inferences and Conclusion

<b> Summary

- 34.7% of students have depression and 65.3% does not
- 33.7% of students have anxiety and 66.3% does not
- 32.7% of students have depression and 67.3% does not
- 5.9% of students seek specialist treament and 94.1% does not
- 82.9% of female students and 17.1%  of male have depression 
- 70.6% of female students and 29.4%  of male have depression 
- 75.8% of female students and 24.2%  of male have depression 
- 83.3% of female students and 16.7%  of male have depression 

<b> Mental health by CGPA

- Students with CGPA between 3-3.49 tend to have depression more than other CGPA level
- Students with CGPA between 3.50-4.00 tend to have anxiety more than other CGPA level
- Students with CGPA between 3-50-4.00 tend to have panic attack more than other CGPA level

<b> Mental health by Age group

- Students within the age group 15-25 tend to have depression, anxiety and panic attack more than other age group

<b> Seeking treatment by Age group

- Students within the age group 19-25 have accessed treatment for depression, anxiety and panic attack more than other age group

<b> Mental health by Year of Study

- Students in year 1 tend to have depression more than other year of study(in this order of year 1-3-2-4)
- Students in year 1 tend to have anxiety more than other year of study(in this order of year 1-2-3-4)
- Students in year 1 tend to have panic attack more than other year of study(in this order of year 1-3-2-4)

<b> Mental health by Marital Status

- Single students tend to have depression, anxiety and panic attack more than married students

<b> Mental health by Department

- Students in Engineering, BCS and BIT tend to have depression, anxiety and panic attack more than students in other departments
